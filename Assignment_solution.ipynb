```python
import json
import re

def main():
    while True:
        try:
            path = input("Please provide the path of input file: ")
            with open(path, 'r') as json_file:
                data = json.loads(json_file.read())
            break
        except:
            print("This is not proper path.")
            continue
    #if data['function'] == 'sample-text-function':
    first_func(data["questions"])


def first_func(questions):

    localVar = {}

    for level_1_data in questions:
        if level_1_data.get('instruction'):
            instructions_part(level_1_data, localVar)
        elif level_1_data.get("text"):
            list_instance = re.findall('.*(\[.*?\]).*', level_1_data["var"])
            if list_instance:
                temp_var = input_var(level_1_data, localVar)
                list_eval = "{0}.append('{1}')".format(level_1_data["var"].split("[")[0], temp_var)
                eval(list_eval, localVar)
            else:
                localVar[level_1_data["var"]] = input_var(level_1_data, localVar)
        elif level_1_data.get("calculated_variable"):
            #print(level_1_data["var"])
            localVar[level_1_data["var"]] = eval(level_1_data["formula"], localVar)

def input_var(data, localVar):
    var = data["var"]

    if data.get("conditions"):
        conditions = data.get("conditions")
        and_cond = ["(" + " and ".join(row) + ")" for row in conditions]
        cond = " or ".join(and_cond)

        while eval(cond, localVar):
            localVar[var] = take_input(data["text"], data.get("options"))
    else:
        # take input
        temp_var = take_input(data["text"], data.get("options"))

    return localVar.get(var) or temp_var

def instructions_part(data, localVar):
    output = data["instruction"]

    if data.get("list_var") and data.get("list_length"):
        for i in range(0, int(data["list_length"])):
            localVar["i"] = i
            args = [eval(x, localVar) for x in data["instruction_var"]]
            output = data["instruction"] % tuple(args)
            print(output)

        return
    elif data.get("instruction_var"):
        args = [localVar[d] for d in data["instruction_var"]]
        output = output % tuple(args)

    print(output)

def take_input(text, options=[]):
    inp, opt = "", ""
    if options:
        opt = "(" + " / ".join(options) + ")"


    inp = input("{0} {1}: ".format(text, opt))

    if options and inp not in options:
        print("Invalid option, This can't be selected. Please select one of these {0}".format(opt))
        inp = take_input(text, options)

    return inp

if __name__ == "__main__":
    main()
```
